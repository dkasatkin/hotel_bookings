# Бронирование отелей #

$\large{Необходимо\: проанализировать\: данные\: о\: бронировании\: отелей,\: содержащиеся\: в\: датасете:}$

* <span style="color:red"><b>Hotel</b></span> – тип отеля (<span style="color:blue">City Hotel</span> или <span style="color:blue">Resort Hotel</span>)  
* <span style="color:red"><b>Is canceled</b></span> – бронирование было отменено -> <span style="color:blue">1</span> или <span style="color:blue">0</span> -> нет; не отмененное считается успешным
* <span style="color:red"><b>Lead time</b></span> – количество дней, прошедших между датой бронирования и датой прибытия  
* <span style="color:red"><b>Arrival full date</b></span> – полная дата прибытия
* <span style="color:red"><b>Arrival date year</b></span> – год прибытия  
* <span style="color:red"><b>Arrival date month</b></span> – месяц прибытия  
* <span style="color:red"><b>Arrival date week number</b></span> – номер недели прибытия
* <span style="color:red"><b>Arrival date day of month</b></span> – день прибытия
* <span style="color:red"><b>Stays in weekend nights</b></span> – количество выходных (<span style="color:blue">суббота</span> или <span style="color:blue">воскресенье</span>), которые гость забронировал для проживания в отеле
* <span style="color:red"><b>Stays in week nights</b></span> – количество дней (с понедельника по пятницу), которые гость забронировал для проживания в отеле
* <span style="color:red"><b>Stays total nights</b></span> – общее число забронированных ночей (сумма двух предыдущих колонок)
* <span style="color:red"><b>Adults</b></span> – число взрослых
* <span style="color:red"><b>Children</b></span> – число детей
* <span style="color:red"><b>Babies</b></span> – число младенцев 
* <span style="color:red"><b>Meal</b></span> – выбранный тип питания
* <span style="color:red"><b>Country</b></span> – страна происхождения клиента
* <span style="color:red"><b>Reserved room type</b></span> – тип зарезервированного номера
* <span style="color:red"><b>Assigned room type</b></span> – тип полученного номера (может отличаться от забронированного)
* <span style="color:red"><b>Customer type</b></span> – тип бронирования
* <span style="color:red"><b>Reservation status</b></span> – значение последнего статуса брони: <span style="color:blue">Canceled</span> - было отменено клиентом; <span style="color:blue">Check-Out</span> - клиент зарегистрировался, но уже покинул отель; <span style="color:blue">No-Show</span> - клиент не зарегистрировался и сообщил администрации отеля причину
* <span style="color:red"><b>Reservation status date</b></span> – дата обновления статуса

```python
# Импортируем библиотеку

import pandas as pd
```

```python
# Загружаем датасет
# Смотрим размер таблицы, тип переменных

booking_df = pd.read_csv('2_bookings.csv', sep=';')
print(booking_df.shape,
      booking_df.dtypes, 
      sep='\n' + '--' * 20 + '\n')
```

```python
# Подгружаем первые 7 строк 

booking_df.head(7)
```

```python
# Приведем названия столбцов в приемлемый вид:
# все колонки должны быть в нижним регистре; 
# вместо пробелов - знаки нижнего подчеркивания

booking_df.columns = booking_df.columns.str.lower()
booking_df.columns = booking_df.columns.str.replace(' ', '_')
booking_df.head()
```

## 1. Укажем топ-5 стран, из которых пользователи совершили наибольшее число успешных бронирований

```python
# здесь необходимо подсчитать число вхождений в каждую страну
# при условии, что бронирование успешное

booking_df.country[booking_df.is_canceled != 1].value_counts()[:5]

# как видим топ-5: PRT -> Португалия
#                  GBR -> Великобритания
#                  FRA -> Франция
#                  ESP -> Испания
#                  DEU -> Германия
```

 ## 2. На сколько ночей в среднем бронируют отели разных типов?

```python
# Находим среднее по формуле:
# (кол-во_ночей_пользователя_1 + ... + кол-во_ночей_пользователя_n) / n;
# или с помощью встроенной функции
# А далее сравниваем значения для разных типов отелей

booking_df.groupby('hotel').mean().stays_total_nights

# как видим в среднем бронируют курортные отели чаще, чем отели в черте города
```

## 3. Иногда тип номера, полученного клиентом, отличается от изначально забронированного. Сколько подобных наблюдений встретилось в датасете?

```python
# Количество таких наблюдений будет равно длине датафрейма, в котором будут записаны данные,
# отобранные по условию: заброннированный тип не равен полученному типу

booking_df[booking_df.assigned_room_type != booking_df.reserved_room_type].shape
```

## 4. На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?

```python
# Здесь нам интересны три столбца: успешна ли бронь, месяц и год
# Рассмотрим для 2016 года

booking_df[(booking_df.is_canceled != 1) & 
           (booking_df.arrival_date_year == 2016)] \
           .groupby('arrival_date_month') \
           .aggregate({'arrival_date_year': 'sum'}) \
           .sort_values('arrival_date_year', ascending=False)
```

```python
# Для 2017 года

booking_df[(booking_df.is_canceled != 1) & 
           (booking_df.arrival_date_year == 2017)] \
           .groupby('arrival_date_month') \
           .aggregate({'arrival_date_year': 'sum'}) \
           .sort_values('arrival_date_year', ascending=False)

# В 2017 году самый популярный месяц сменился. В 2016 -> октябрь, в 2017 ->  май
```

```python
# Здесь еще посмотрим, на какой месяц бронирования отеля типа City Hotel
# отменялись чаще всего в каждый из периодов (2015? 2016? 2017?)

booking_df[(booking_df.is_canceled == 1) & 
          (booking_df.hotel == 'City Hotel')] \
                                             .groupby(['arrival_date_year', 'arrival_date_month']) \
                                             .aggregate({'is_canceled': 'sum'}) \
                                             .sort_values(['arrival_date_year', 'is_canceled'], ascending=False)

# 2017 -> Май
# 2016 -> Октябрь
# 2015 -> Сентябрь
```

## 5. Какая из числовых характеристик adults, children или babies имеет наибольшее среднее значение?

```python
# Выберем три столбца из датафрейма 
# и посмотрим среднее для каждого столбца

booking_df[['children', 'adults', 'babies']].mean()
```

## 6. Если объединить детей всех возрастов, то для отелей какого типа среднее значение новой переменной окажется наибольшим?

```python
# Создадим новый столбец в нашем датафрейме
# и запишем в него сумму столбцов children и babies

booking_df['sum_of_child'] = booking_df['children'] + booking_df['babies']

# После сгруппируем данные по типу отеля,
# и посчитаем среднее для каждого значения

booking_df.groupby('hotel') \
          .aggregate({'sum_of_child': 'mean'})

# В среднем в курортных отелях количество детей больше
```

## 7. Какое отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах и среди какой группы этот показатель выше: у врослых с детьми или без детей?

```python
# Churn rate (отток, коэффициент оттока) – это процент подписчиков 
# (например, на push-уведомления от сайта), которые отписались от 
# канала коммуникации, отказались от услуг сервиса в течение определенного
# периода времени. Иными словами, представляет собой отношение количества 
# ушедших пользователей к общему количеству пользователей, выраженное в процентах.
```

```python
# booking_df['has_kids'].loc[booking_df['sum_of_child'] >= 1] = True
booking_df.loc[booking_df['sum_of_child'] > 0, 'has_child'] = True
booking_df.loc[booking_df['sum_of_child'] == 0.0, 'has_child'] = False
```

```python
booking_df[booking_df['is_canceled'] == 1].has_child.value_counts()
```

```python
booking_df.has_child.value_counts()
```

```python
# Считает отношение churn rate в категории "без детей"

40961 / 110058 * 100
```

```python
# Считаем отношение churn rate в категории "с детьми"

3259 / 9332 * 100
```

```python
# Альтернативное решение
booking_df['has_kids'] = booking_df.sum_of_child > 0
```

```python
without_child = (booking_df.query('is_canceled == 1 & has_kids == False').shape[0] /
booking_df.query('has_kids == False').shape[0]) * 100
```

```python
with_child = (booking_df.query('is_canceled == 1 & has_kids == True').shape[0] /
booking_df.query('has_kids == True').shape[0]) * 100
```

```python
print(with_child, without_child)
```
